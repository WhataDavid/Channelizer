```python
import numpy as np
import matplotlib.pyplot as plt
from dtw import dtw
from scipy.spatial.distance import euclidean

def draw_dtw(signal1, signal2,signal1_name,signal2_name,):

    # 使用DTW算法计算距离
    distance, cost_matrix, acc_cost_matrix, path = dtw(signal1.reshape(-1, 1), signal2.reshape(-1, 1), dist=euclidean)

    # 打印DTW距离
    # print(f"DTW value: {distance}")

    # 可视化结果
    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1)
    plt.plot(signal1, label=signal1_name)
    plt.plot( signal2, label=signal2_name)
    plt.legend()
    plt.title("compare "+signal1_name+" with "+signal2_name)

    plt.subplot(1, 2, 2)
    plt.imshow(acc_cost_matrix.T, origin='lower', cmap='gray', interpolation='nearest')
    plt.plot(path[0], path[1], 'w')
    plt.title('Accumulated cost matrix and optimal path')
    # 累积成本矩阵和最优路径
    plt.tight_layout()
    plt.show()


origin = np.loadtxt("txt/subcspfb_2channel_all_correct.txt")
cspfb = np.loadtxt("txt/cspfb_20channel_all_correct.txt")
opfb2x = np.loadtxt("txt/ospfb2x_20channel_all_correct.txt")
opfb4_3x = np.loadtxt("txt/ospfb43x_20channel_all_correct.txt")

origin = (origin - np.min(origin)) / (np.max(origin) - np.min(origin))
cspfb = (cspfb - np.min(cspfb)) / (np.max(cspfb) - np.min(cspfb))
opfb2x = (opfb2x - np.min(opfb2x)) / (np.max(opfb2x) - np.min(opfb2x))
opfb4_3x = (opfb4_3x - np.min(opfb4_3x)) / (np.max(opfb4_3x) - np.min(opfb4_3x))
```

```python
print(origin.shape)
print(cspfb.shape)
print(opfb2x.shape)
print(opfb4_3x.shape)

plt.figure(figsize=(15,7))
plt.plot(origin,label="origin")
plt.plot(cspfb,label="cspfb")
plt.legend()
plt.show()
draw_dtw(origin,cspfb,"origin","cspfb")
```

```python
plt.figure(figsize=(15,7))
plt.plot(origin,label="origin")
plt.plot(opfb2x,label="opfb 2x")
plt.legend()
plt.show()
draw_dtw(origin,opfb2x,"origin","opfb 2x")
```

```python
plt.figure(figsize=(15,7))
plt.plot(origin,label="origin")
plt.plot(opfb4_3x,label="opfb 4/3x")
plt.legend()
plt.show()
draw_dtw(origin,opfb4_3x,"origin","opfb 4/3x")
```
